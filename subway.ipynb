```python
import matplotlib.pyplot as plt # 다양한 형태의 그래프 그릴 수 있는 라이브러리
import seaborn as sns # 고급 시각화 라이브러리(통계적 그래프)
import pandas as pd # 데이터 조작과 분석을 위한 라이브러리(데이터프레임 구조를 사용해서 데이터 처리)

# 함글 폰트 설정
plt.rcParams['font.family'] = 'AppleGothic' # 맥에서 사용할 수 있는 한글 폰트

# 마이너스 폰트 설정
plt.rcParams['axes.unicode_minus'] = False # 마이너스 기호와 한글 폰트와의 충돌 방지 
```

```python
import chardet

# CSV 파일 경로
file_path = '/Users/baejuyeon/Downloads/서울교통공사_지하철혼잡도정보_20230331.csv'

# 파일의 인코딩 감지
with open(file_path, 'rb') as f:
    result = chardet.detect(f.read())

encoding = result['encoding']

# 감지된 인코딩으로 CSV 파일 불러오기
df = pd.read_csv(file_path, encoding=encoding)

# 소수점 두 자리로 출력 옵션 설정
pd.options.display.float_format = '{:.1f}'.format

# 데이터 확인
df.head()
```

```python
# 시간대 컬럼들 선택
time_columns = df.columns[6:]

# 데이터 long format으로 변환
df_melted = df.melt(id_vars=['연번', '요일구분', '호선', '역번호', '출발역', '상하구분'], value_vars=time_columns, var_name='시간', value_name='혼잡도')

# 시간 컬럼을 숫자로 변환 (예: '5시30분' -> 5.5)
df_melted['시간_시'] = df_melted['시간'].str.extract('(\d+)시').astype(int)
df_melted['시간_분'] = df_melted['시간'].str.extract('(\d+)분').astype(int)
df_melted['시간'] = df_melted['시간_시'] + df_melted['시간_분'] / 60

# 시간 컬럼 변환에 사용한 임시 컬럼 제거
df_melted.drop(columns=['시간_시', '시간_분'], inplace=True)

# 요일별 시간대별 복잡도 시각화
plt.figure(figsize=(12, 6))
sns.lineplot(data=df_melted, x='시간', y='혼잡도', hue='요일구분')
plt.title('요일별 시간대별 혼잡도')
plt.xlabel('시간')
plt.ylabel('혼잡도')
plt.legend(title='요일구분')
plt.grid(True)
plt.show()

# 요일별 호선별 시간대별 복잡도 시각화
plt.figure(figsize=(12, 6))
sns.lineplot(data=df_melted, x='시간', y='혼잡도', hue='호선', style='요일구분')
plt.title('요일별 호선별 시간대별 혼잡도')
plt.xlabel('시간')
plt.ylabel('혼잡도')
plt.legend(title='호선')
plt.grid(True)
plt.show()

# 요일별 호선별 출발역별 상하구분별 시간대별 복잡도 시각화
plt.figure(figsize=(12, 6))
sns.lineplot(data=df_melted, x='시간', y='혼잡도', hue='출발역', style='상하구분', size='호선', palette='tab20')
plt.title('요일별 호선별 출발역별 상하구분별 시간대별 혼잡도')
plt.xlabel('시간')
plt.ylabel('혼잡도')
plt.legend(title='출발역')
plt.grid(True)
plt.show()
```

```python
# 시간대 컬럼들 선택
time_columns = df.columns[6:]

# 데이터 long format으로 변환
df_melted = df.melt(id_vars=['연번', '요일구분', '호선', '역번호', '출발역', '상하구분'], value_vars=time_columns, var_name='시간', value_name='혼잡도')

# 시간 컬럼을 숫자로 변환 (예: '5시30분' -> 5.5)
df_melted['시간_시'] = df_melted['시간'].str.extract('(\d+)시').astype(int)
df_melted['시간_분'] = df_melted['시간'].str.extract('(\d+)분').astype(int)
df_melted['시간'] = df_melted['시간_시'] + df_melted['시간_분'] / 60

# 시간 컬럼 변환에 사용한 임시 컬럼 제거
df_melted.drop(columns=['시간_시', '시간_분'], inplace=True)

# 호선별로 요일 구분(평일, 주말) 나누어서 시간별 혼잡도 시각화
unique_lines = df_melted['호선'].unique()
num_lines = len(unique_lines)
fig, axes = plt.subplots(num_lines, 1, figsize=(12, 6 * num_lines), sharex=True)

for i, line in enumerate(unique_lines):
    ax = axes[i] if num_lines > 1 else axes
    sns.lineplot(data=df_melted[df_melted['호선'] == line], x='시간', y='혼잡도', hue='요일구분', ax=ax)
    ax.set_title(f'{line} 시간대별 혼잡도')
    ax.set_xlabel('시간')
    ax.set_ylabel('혼잡도')
    ax.legend(title='요일구분')
    ax.grid(True)

plt.tight_layout()
plt.show()
```
